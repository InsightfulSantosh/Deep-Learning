  - Multi-layer perception is also known as MLP. It is fully connected dense layers, which transform any input dimension to the desired dimension. A multi-layer perception is a neural network that has multiple layers. To create a neural network we combine neurons together so that the outputs of some neurons are inputs of other neurons.
  - MLPs serve as powerful tools in AI due to their capacity to learn from data, generalize to unseen examples, and handle complex patterns. They form a cornerstone in the world of neural networks, enabling us to tackle a wide range of real-world challenges.

* **Lets create the network more functional**
  

```python
Inputs = [1.5, 6.2, 3.2, 2.1]

Weights = [[0.5, 0.9, 0.4, -0.5],
           [2.2, 1.9, 1.1, -0.8],
           [0.1, 0.3, -0.4, 0.5]]

Biases = [1.2, 0.8, 1.1]

layer_outputs = []

for neuron_weights, neuron_bias in zip(Weights, Biases):
    neuron_output = 0
    for n_input, weight in zip(Inputs, neuron_weights):
        neuron_output += n_input*weight
    neuron_output += neuron_bias
    layer_outputs.append(neuron_output)
    
print(layer_outputs)
```

* **Now we can add Activation Function into the network (Sigmoid and ReLU)**

```python
nnfs.init()


class Layer_dense:
    
    #init save our number of inputs, neurons and weights.
    
    def __init__ (self, n_inputs, n_neurons):
        
        # we can normalize the weights for avoiding (Vanish and Explode)
        self.weights = 0.10*np.random.randn(n_inputs, n_neurons)
        
        #Add biases from number of neurons and set all of them 0
        self.biases = np.zeros((1, n_neurons))
        
    # The Feedforward processing    
    def forward(self, inputs):
        
        #Inputs*Weights + Biases
        self.output = np.dot(inputs, self.weights) + self.biases
        
        
#Actication Function ReLU

class Activation_ReLU:
    def forward(self, inputs):
        self.output = np.maximum (0, inputs)
        
        
        
#Also we can create Softmax Activation function

class Activation_Softmax:
    def forward(self, inputs):
        exp_values = np.exp(inputs - np.max(inputs, axis=1, keepdims=True))
        probabilities = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        self.output = probabilities
        

        
#Loss function
class Loss:
    def calculate(self, output, y):
        sample_losses = self.forward(output, y)
        data_loss = np.mean(sample_losses)
        return data_loss
    
class Loss_Cat_CrossEntropy (Loss):
    def forward(self, y_pred, y_ture):
        y_true = y
        samples = len(y_pred)
        #The numpy clip function is used to limit the upper and lower values of an array
        y_pred_clipped = np.clip(y_pred, 1e-7, 1-1e-7)
        
        if len(y_true.shape)==1:
            correct_confidences = y_pred_clipped[range(samples), y_ture]
         #Encoding the label   
        elif len(y_ture.shape) ==2:
            correct_confidences = np.sum(y_pred_clipped*y_true, axis=1)
            
        #Now it's time to calculate negative log_likelihood that Use sample_losses, data_loss 
        negative_log_likelihood = -np.log(correct_confidences)
        return negative_log_likelihood

        
X, y = spiral_data(samples= 100, classes= 3)

dense1 = Layer_dense(2,3)
activation1 = Activation_ReLU()

dense2 = Layer_dense(3,3)
activation2 = Activation_Softmax()

dense1.forward(X)
activation1.forward(dense1.output)
dense2.forward(activation1.output)
activation2.forward(dense2.output)

print("The Output of Softmax Activation is : \n", activation2.output[:5], "\n")

loss_function = Loss_Cat_CrossEntropy()
loss = loss_function.calculate(activation2.output, y)

print("Loss is: \n", loss)
```

* **Let's create neural network with Tensorflow**

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data() 
```

```python
# Cast the records into float values 
x_train = x_train.astype('float32') 
x_test = x_test.astype('float32') 
  
# normalize image pixel values by dividing  
# by 255 
gray_scale = 255
x_train /= gray_scale 
x_test /= gray_scale
 
print("Feature matrix:", x_train.shape) 
print("Target matrix:", x_test.shape) 
print("Feature matrix:", y_train.shape) 
print("Target matrix:", y_test.shape) 
```

```python
fig, ax = plt.subplots(5, 5)
k = 0
for i in range(5): 
    for j in range(5): 
        ax[i][j].imshow(x_train[k].reshape(28, 28),  
                        aspect='auto') 
        k += 1
plt.show() 
```

```python
model = Sequential([ 
    
    # reshape 28 row * 28 column data to 28*28 rows 
    Flatten(input_shape=(28, 28)), 
    
      # dense layer 1 
    Dense(256, activation='sigmoid'),   
    
    # dense layer 2 
    Dense(128, activation='sigmoid'),  
    
      # output layer 
    Dense(10, activation='sigmoid'),   
]) 
```

```python
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy']) 
```

```python
model = model.fit(x_train, y_train, epochs=100,  
          batch_size=2000,  
          validation_split=0.2) 
```

```python
training_loss = model.history['loss']
val_loss = model.history['val_loss']
training_acc = model.history['accuracy']
val_acc = model.history['val_accuracy']
```

```python
epoch = range(1,101)
plt.figure(figsize=(10,5), dpi=200)
plt.plot(epoch, training_loss, 'r--', color= 'navy')
plt.plot(epoch, val_loss, '--bo',color= 'orangered',linewidth = '2.5', label='line with marker')
plt.legend(['Training Loss', 'Val Loss'])
plt.title('Number of epochs & Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(np.arange(0,105,5))
plt.show();
```

```python
plt.figure(figsize=(10,5), dpi=200)
plt.plot(epoch, training_acc, 'r--', color= 'navy')
plt.plot(epoch, val_acc, '--bo',color= 'orangered',linewidth = '2.5', label='line with marker')
plt.legend(['Training Acc', 'Val Acc'])
plt.title('Number of epochs & ACC')
plt.xlabel('Epoch')
plt.ylabel('Acc')
plt.xticks(np.arange(0,105,5))
plt.show();
```

* **Let's create neural network with Pytorch**

* **[Table of Content](#0)**

```python
# Step 1: Load and preprocess the data
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, transform=transform, download=True)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
```

```python
# Step 2: Define the MLP architecture
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(28 * 28, 128)  # Input layer
        self.fc2 = nn.Linear(128, 64)      # Hidden layer
        self.fc3 = nn.Linear(64, 10)       # Output layer
        
    def forward(self, x):
        x = x.view(-1, 28 * 28)  # Flatten the input
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
# Step 3: Train the model
model = MLP()
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)
```

```python
for epoch in range(10):  # You can adjust the number of epochs
    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

print("Training complete!")

print(loss)
# Now you can use the trained model for digit classification.
```
