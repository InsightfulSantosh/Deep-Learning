### Recurrent Neural Networks (RNN): Detailed Overview

#### 1. **What is RNN?**
Recurrent Neural Networks (RNNs) are a class of artificial neural networks designed to recognize patterns in sequences of data. Unlike traditional neural networks, which assume input data points are independent of each other, RNNs are specifically designed to handle sequential data by considering the temporal relationships between the inputs.

RNNs maintain a "memory" of previous inputs by using internal hidden states that allow them to capture information from previous time steps. This makes them ideal for tasks where the order of the data matters, such as time series prediction, language modeling, and speech recognition.


### **Types of RNN:**

image.png

### 1. **One to One (Basic RNN)**

- **Definition**: A **one-to-one** RNN is the simplest form of an RNN. It processes one input and produces one output, making it similar to conventional feed-forward neural networks.
  
- **Use Case**: This architecture is suitable for tasks where the input and output are fixed in size and theres no need to process sequential data. Examples include image classification, where a single image is classified into a category.

- **Example**:
  - **Input**: A single image.
  - **Output**: A label for the image (e.g., "cat").
  
  Here, theres no temporal sequence involvedjust one input leading to one output.

---

### 2. **One to Many (Image Captioning)**

- **Definition**: In the **one-to-many** architecture, the model receives a single input but produces a sequence of outputs over time.
  
- **Use Case**: This architecture is useful for tasks like **image description or captioning**, where a single input (an image) results in multiple outputs (a sentence describing the image). The output sequence is generated over several time steps.

- **Example**:
  - **Input**: An image.
  - **Output**: A sequence of words describing the image.
  
  For instance, given an image of a dog playing in a park, the output could be: `["A", "dog", "is", "playing", "in", "the", "park"]`.

- **How It Works**:
  - The RNN takes in the image (after processing it through a convolutional neural network, CNN), and then generates a sequence of words that form a sentence.

---

### 3. **Many to One (Sentiment Analysis)**

- **Definition**: The **many-to-one** RNN architecture processes a sequence of inputs and produces a single output.
  
- **Use Case**: This type is commonly used for tasks where the input is sequential, but the output is a single value, like **sentiment analysis**. Here, a sequence of words (input) leads to a single sentiment classification (output).

- **Example**:
  - **Input**: A sequence of words from a movie review.
  - **Output**: A single sentiment (e.g., "positive" or "negative").
  
  For instance, given the sentence The movie was fantastic and thrilling, the output could be a label: **positive**.

- **How It Works**:
  - The RNN reads the entire input sequence (e.g., a sentence or document) and produces a final output after processing the sequence step by step.

---

### 4. **Many to Many (Language Translation)**

- **Definition**: The **many-to-many** RNN architecture processes a sequence of inputs and produces a sequence of outputs. This is typically used in tasks that require the retention of sequence structure, like **language translation** or **speech recognition**.
  
- **Use Case**: **Language translation** is a perfect example of this architecture. The model takes a sentence in one language as input and produces a corresponding sentence in another language as output.

- **Example**:
  - **Input**: A sentence in English, e.g., "How are you?"
  - **Output**: The corresponding sentence in Spanish, e.g., "Cmo ests?"

- **How It Works**:
  - The RNN reads the entire input sequence and generates an output sequence, step by step, while maintaining context through hidden states. This process can be enhanced using an **encoder-decoder** architecture (like in machine translation tasks), where one RNN encodes the input sequence, and another decodes it to generate the output.

#### 2. **Varriants of RNNs**
There are various varriants of RNN architectures, each designed for different types of sequence processing tasks:

- **Vanilla RNN**: The basic form of RNN with a single hidden state. This architecture is limited by issues like vanishing and exploding gradients.

- **Long Short-Term Memory (LSTM)**: LSTMs are a type of RNN designed to capture long-term dependencies. They use gates (input, forget, output) to control the flow of information, allowing the model to learn which parts of the sequence to remember and which to forget.

- **Gated Recurrent Units (GRU)**: GRUs are a simpler version of LSTMs with fewer gates (reset and update gates), making them computationally faster but still effective at capturing dependencies in sequence data.

- **Bidirectional RNN (Bi-RNN)**: This type of RNN processes the input sequence in both forward and backward directions, allowing it to capture information from both past and future contexts.

- **Deep RNN**: These are RNNs with multiple layers of hidden units, making them more powerful but also more prone to vanishing gradients.

---------
#### 3. **RNN Architecture**
A typical RNN architecture consists of:

- **Input Layer**: Receives the input sequence. Each element in the sequence is fed one by one.
  
- **Hidden Layer(s)**: This layer maintains the memory of the RNN. At each time step, the hidden state is updated based on the current input and the previous hidden state. Mathematically, this can be represented as:
  $$
  h_t = \tanh(W_h \cdot h_{t-1} + W_x \cdot x_t + b)
  $$
  where:
  - $( h_t )$ is the hidden state at time step $( t )$
  - $( h_{t-1} )$ is the hidden state from the previous time step
  - $( W_h )$ and $( W_x )$ are weight matrices
  - $( x_t )$ is the current input
  - $( b )$ is the bias

- **Output Layer**: Produces the final prediction/output for each time step. The output could either be a sequence of values (for sequence-to-sequence tasks) or a single value after processing the entire input (for tasks like sentiment analysis).

--------
#### 4. **How RNN Works**
RNNs work by maintaining a hidden state that gets updated at each time step. Here's a step-by-step explanation:

1. **Input**: The input sequence is fed into the network one time step at a time. Each element of the sequence influences the hidden state of the network.

2. **Hidden State Update**: At each time step, the hidden state is updated based on the current input and the previous hidden state. This update allows the RNN to retain information about the sequence.

3. **Output**: The output at each time step is generated based on the hidden state. Depending on the task, this output can be used immediately (e.g., in sequence-to-sequence tasks) or after the entire sequence has been processed (e.g., for classification tasks).

4. **Training**: RNNs are typically trained using backpropagation through time (BPTT), a variant of backpropagation that unfolds the network over time and adjusts weights to minimize the loss.

-------
#### 5. **Uses of RNN**
RNNs are commonly used in tasks that involve sequential data, including:

- **Natural Language Processing (NLP)**: Language modeling, machine translation, text generation, and sentiment analysis.
  
- **Speech Recognition**: Recognizing spoken words and converting them to text.
  
- **Time Series Prediction**: Predicting future values in a sequence based on past observations (e.g., stock prices, weather forecasting).
  
- **Video Analysis**: Capturing temporal dynamics in video frames.
  
- **Music Generation**: Creating sequences of notes in a musical piece.

#### 6. **Limitations of RNNs**
RNNs, especially Vanilla RNNs, suffer from several key limitations:

- **Vanishing and Exploding Gradient Problems**: During backpropagation through time, gradients can become extremely small (vanish) or very large (explode), leading to poor learning. This makes it difficult for Vanilla RNNs to learn long-term dependencies in sequences.

- **Short-Term Memory**: Due to the vanishing gradient problem, RNNs are better suited for short-term dependencies, and they struggle to remember information from far back in the sequence.

- **Sequential Processing**: RNNs process input data sequentially, which can be computationally expensive, especially for long sequences.

- **Difficulty in Training**: Training RNNs is generally harder and more time-consuming compared to feedforward neural networks due to the complexity of handling sequences.

-------
#### 7. **Problems with RNNs**
There are several inherent problems with RNNs:

- **Vanishing Gradient Problem**: As the sequence length increases, the gradients propagated backward during training become smaller and smaller. This makes it hard for the network to learn long-term dependencies because the influence of earlier time steps diminishes.

- **Exploding Gradient Problem**: Conversely, gradients can sometimes become too large, causing the network weights to update in a way that results in numerical instability or failure to converge.

- **Limited Long-Term Memory**: Even though RNNs have a hidden state to capture sequence information, they are typically limited in their ability to retain information over long time steps, which is where LSTMs and GRUs come into play.

- **Slow Training**: Training RNNs can be slow, especially for long sequences, because each time step must be processed in sequence (unlike CNNs, which can process input in parallel).

- **High Complexity in Sequence Generation**: Generating sequences, such as sentences or music, with RNNs can be challenging, especially when long-term structure and coherence are required.

### **Step-by-Step Derivation and Working of RNN Model**

#### 1. **RNN Forward Propagation**

An RNN processes input sequentially by maintaining a hidden state across time steps. At each time step $( t )$, the RNN updates its hidden state based on the current input and the previous hidden state. This is the core idea behind the RNN.

For simplicity, let's assume that the RNN is a single-layer network. The equations governing the forward pass of an RNN can be derived as follows:

- Let the input at time step $( t )$ be $( x_t \in \mathbb{R}^n)$, where $( n )$ is the dimension of the input vector.
- Let the hidden state at time step $( t )$ be $( h_t \in \mathbb{R}^m )$, where $( m )$ is the number of hidden units.
- Let the output at time step $( t )$ be $( o_t \in \mathbb{R}^k )$, where $( k )$ is the dimension of the output vector.

The key equations for the forward propagation are:

1. **Hidden state update**:
   $$
   h_t = \tanh(W_h h_{t-1} + W_x x_t + b_h)
   $$
   where:
   - $( W_h \in \mathbb{R}^{m \times m} )$ is the weight matrix for the hidden state.
   - $( W_x \in \mathbb{R}^{m \times n} )$ is the weight matrix for the input.
   - $( b_h \in \mathbb{R}^{m} )$ is the bias vector.
   - $( \tanh )$ is the activation function applied element-wise.

2. **Output computation**:
   $$
   o_t = W_o h_t + b_o
   $$
   where:
   - $( W_o \in \mathbb{R}^{k \times m} )$ is the weight matrix for the output.
   - $( b_o \in \mathbb{R}^{k} )$ is the bias vector for the output.

The RNN processes the input sequence step by step, updating the hidden state and producing an output at each step.

#### 2. **Loss Function**

Assume the task is to predict a target sequence $( \{ y_1, y_2, ..., y_T \} )$, where $( y_t )$ is the true output at time step $( t )$, and the model's output is $( \hat{y}_t )$.

To quantify the difference between the predicted output and the true output, we use a loss function, commonly the **mean squared error (MSE)** for regression or **cross-entropy loss** for classification.

For classification, the cross-entropy loss at each time step \( t \) is given by:
$$
\mathcal{L}_t = -\sum_{i=1}^{k} y_t^{(i)} \log(\hat{y}_t^{(i)})
$$
where $( \hat{y}_t^{(i)} )$ is the predicted probability of class $( i )$ at time step $( t )$, and $( y_t^{(i)} )$ is the one-hot encoded true label.

The total loss for the entire sequence (of length $( T )$) is:
$$
\mathcal{L} = \frac{1}{T} \sum_{t=1}^{T} \mathcal{L}_t
$$

#### 3. **Backpropagation Through Time (BPTT)**

To minimize the loss, the weights in the RNN are updated using gradient-based optimization methods such as **stochastic gradient descent (SGD)**. The gradients are computed using a process called **Backpropagation Through Time (BPTT)**, which is a modification of the standard backpropagation algorithm to handle sequences.

The main idea is to **unroll** the RNN over time, treating each time step as a separate layer in a deep network. Let's break it down:

1. **Unrolling the RNN**:
   An RNN can be unrolled into a feedforward network by considering each time step as a separate layer, but all the layers share the same weights $( W_h, W_x, W_o )$. This allows the gradients to propagate backward through time.

2. **Gradient of the Loss**:
   To compute the gradients with respect to the parameters $( W_h, W_x, W_o, b_h, b_o )$, we use the chain rule.

   The loss at each time step $( t )$ depends not only on the hidden state $( h_t )$ but also on the previous hidden states $( h_{t-1}, h_{t-2}, \dots, h_1 )$. So, the gradient with respect to the weights involves summing the gradients over all time steps:
  $$
   \frac{\partial \mathcal{L}}{\partial W_h} = \sum_{t=1}^{T} \frac{\partial \mathcal{L}_t}{\partial h_t} \cdot \frac{\partial h_t}{\partial W_h}
 $$

   Similarly, we compute the gradients for $( W_x, W_o, b_h, b_o )$.

3. **Gradient Updates**:
   Once the gradients are computed, the weights are updated using gradient descent:
   $$
   W_h \leftarrow W_h - \eta \frac{\partial \mathcal{L}}{\partial W_h}
   $$
   $$
   W_x \leftarrow W_x - \eta \frac{\partial \mathcal{L}}{\partial W_x}
   $$
   where $( \eta )$ is the learning rate.

#### 4. **Example of Training an RNN**

Lets consider a simple example where we train an RNN to predict the next number in a sequence. Suppose we have the sequence $( \{ 0, 1, 2, 3, 4 \} )$ and the task is to predict the next number. We use a **mean squared error (MSE)** loss function for this regression task.

- Input: $( x_1 = 0, x_2 = 1, x_3 = 2, x_4 = 3, x_5 = 4 )$
- Target Output: $( y_1 = 1, y_2 = 2, y_3 = 3, y_4 = 4, y_5 = 5 )$

1. **Forward Pass**:
   For each time step $( t )$, compute the hidden state and the output:
   $$
   h_1 = \tanh(W_h h_0 + W_x x_1 + b_h)
   $$
   $$
   o_1 = W_o h_1 + b_o
   $$
   Continue this for all time steps.

2. **Loss Computation**:
   Compute the MSE loss for each time step:
   $$
   \mathcal{L}_t = (y_t - o_t)^2
   $$
   The total loss is the average over all time steps.

3. **Backward Pass (BPTT)**:
   Compute the gradients of the loss with respect to the weights using the chain rule. The gradient for a parameter at time step $( t )$ depends on the gradients from future time steps, requiring the summation of gradients over time.

4. **Weight Updates**:
   Update the weights using the computed gradients:
   $$
   W_h \leftarrow W_h - \eta \frac{\partial \mathcal{L}}{\partial W_h}
   $$
   This is done for all parameters $( W_h, W_x, W_o, b_h, b_o )$.

---

#### 5. **Summary of Key Equations**

- Hidden state update:
  $$
  h_t = \tanh(W_h h_{t-1} + W_x x_t + b_h)
  $$
  
- Output computation:
  $$
  o_t = W_o h_t + b_o
  $$

- Loss function (for classification):
  $$
  \mathcal{L}_t = -\sum_{i=1}^{k} y_t^{(i)} \log(\hat{y}_t^{(i)})
  $$

- Total loss:
  $$
  \mathcal{L} = \frac{1}{T} \sum_{t=1}^{T} \mathcal{L}_t
  $$

- Weight updates:
  $$
  W_h \leftarrow W_h - \eta \frac{\partial \mathcal{L}}{\partial W_h}
  $$

This covers the mathematical derivation of RNNs, including the forward pass, loss computation, and backpropagation through time for weight updates.

Let's walk through the step-by-step training process of an RNN on a **textual data example** using a simple sentence for character-level prediction. We'll demonstrate the RNN's forward pass, loss calculation, backpropagation through time (BPTT), and weight updates.

### **Example: Character-level Text Prediction**

Suppose we have the sentence:

```
"hello"
```

The goal is to train an RNN to predict the next character in the sequence. So, the input-output pairs for training will be:

- Input: `"h"`, `"e"`, `"l"`, `"l"`, `"o"`
- Target Output: `"e"`, `"l"`, `"l"`, `"o"`, `"<end>"`

We'll represent the characters using **one-hot encoding**.

#### 1. **One-hot Encoding**
Let's assign each character an index:
- $( h \rightarrow [1, 0, 0, 0, 0] )$
- $( e \rightarrow [0, 1, 0, 0, 0] )$
- $( l \rightarrow [0, 0, 1, 0, 0] )$
- $( o \rightarrow [0, 0, 0, 1, 0] )$
- $( <end> \rightarrow [0, 0, 0, 0, 1] )$

Each character is now a vector of size 5.

#### 2. **RNN Architecture Setup**

We will use:
- Input size $( n = 5 )$ (size of one-hot vectors)
- Hidden size $( m = 4 )$ (arbitrary hidden state size)
- Output size $( k = 5 )$ (same as the input size because we predict the next character)

The RNN equations are:

1. **Hidden state update**:
   $$
   h_t = \tanh(W_h h_{t-1} + W_x x_t + b_h)
   $$
   where:
   - $( h_t \in \mathbb{R}^4 )$ is the hidden state.
   - $( W_h \in \mathbb{R}^{4 \times 4} ), ( W_x \in \mathbb{R}^{4 \times 5} )$ are the weight matrices.
   - $( b_h \in \mathbb{R}^{4} )$ is the bias vector.

2. **Output computation**:
   $$
   o_t = W_o h_t + b_o
   $$
   where:
   - $( o_t \in \mathbb{R}^{5} )$ (since we need to predict the next character).
   - $( W_o \in \mathbb{R}^{5 \times 4} )$ is the weight matrix for the output.

3. **Softmax Layer**: To interpret the output as a probability distribution over the characters, we apply the softmax function:
   $$
   \hat{y}_t^{(i)} = \frac{\exp(o_t^{(i)})}{\sum_{j=1}^{5} \exp(o_t^{(j)})}
   $$
   where $( \hat{y}_t^{(i)} )$ is the predicted probability of character $( i )$ at time step $( t )$.

#### 3. **Forward Propagation**

Lets compute the forward pass for the input sequence "hello".

- **At $( t = 1 )$, input: $( x_1 = \text{"h"} )$ (one-hot vector of "h")**:
  - Hidden state: $( h_1 = \tanh(W_h h_0 + W_x x_1 + b_h) )$.
  - Compute output: $( o_1 = W_o h_1 + b_o )$.
  - Apply softmax to get predicted probabilities: $( \hat{y}_1 )$.

- **At $( t = 2 )$, input: $( x_2 = \text{"e"} )$ (one-hot vector of "e")**:
  - Hidden state: $( h_2 = \tanh(W_h h_1 + W_x x_2 + b_h) )$.
  - Compute output: $( o_2 = W_o h_2 + b_o )$.
  - Apply softmax to get predicted probabilities: $( \hat{y}_2 )$.

Continue this process for $( t = 3, 4, 5 )$ with inputs "l", "l", "o".

#### 4. **Loss Function (Cross-Entropy Loss)**

The loss at each time step $( t )$ is computed using **cross-entropy**. Suppose the true target at $( t = 1 )$ is "e", then the cross-entropy loss is:
$$
\mathcal{L}_1 = -\log(\hat{y}_1^{(\text{index of 'e'})})
$$
We repeat this for each time step and calculate the total loss:
$$
\mathcal{L} = \frac{1}{T} \sum_{t=1}^{T} \mathcal{L}_t
$$
where $( T = 5 )$ is the total number of time steps (i.e., the length of the input sequence).

#### 5. **Backpropagation Through Time (BPTT)**

Now, we compute the gradients of the loss with respect to the model parameters $( W_h, W_x, W_o, b_h, b_o )$ using backpropagation through time.

1. **Gradients for output layer**:
   - For each time step $( t )$, the gradient of the loss with respect to the output $( o_t )$ is:
     $$
     \frac{\partial \mathcal{L}_t}{\partial o_t} = \hat{y}_t - y_t
     $$
     where $( y_t )$ is the true one-hot vector at time step $( t )$.

2. **Gradients for hidden state**:
   - The gradient of the loss at time step $( t )$ also depends on the hidden states from future time steps because the hidden states are interconnected. So, we accumulate the gradients for the hidden state:
$$
\frac{\partial \mathcal{L}_t}{\partial h_t} = W_o^T \frac{\partial \mathcal{L}_t}{\partial o_t} + W_h^T \frac{\partial \mathcal{L}_{t+1}}{\partial h_{t+1}}
$$
     This recursive dependence is why its called **backpropagation through time**.

3. **Gradient for weight matrices**:
   - The gradients for $( W_h )$, $( W_x )$, and $( W_o )$ are computed by summing the contributions from all time steps:
$$
\frac{\partial \mathcal{L}}{\partial W_h} = \sum_{t=1}^{T} \frac{\partial \mathcal{L}_t}{\partial h_t} \cdot \frac{\partial h_t}{\partial W_h}
$$

Similarly for $( W_x )$ and $( W_o )$.

#### 6. **Weight Updates**

After computing the gradients, we update the weights using **stochastic gradient descent (SGD)** or another optimizer:
$$
W_h \leftarrow W_h - \eta \frac{\partial \mathcal{L}}{\partial W_h}
$$
where $( \eta )$ is the learning rate. This is done for all the weight matrices $( W_h, W_x, W_o, b_h, b_o )$.

---

### Summary of the Process for Textual Data:

1. **Input**: The RNN receives a sequence of one-hot encoded vectors representing characters (e.g., "hello").
2. **Hidden State Update**: At each time step, the RNN updates its hidden state based on the current input and the previous hidden state.
3. **Output**: The RNN produces an output at each time step, which is passed through a softmax layer to predict the probability distribution of the next character.
4. **Loss**: The cross-entropy loss is computed based on the true target sequence (e.g., "elloh").
5. **Backpropagation Through Time**: Gradients of the loss are computed and propagated backward through time, allowing the model to learn dependencies between characters.
6. **Weight Updates**: The weights are updated using the computed gradients to minimize the loss.

This iterative process continues, allowing the RNN to learn to predict the next character in sequences of text.

**Here are the activation functions with the appropriate mathematical expressions:**

### 1. **Sigmoid Function** ($\sigma(x)$)

- **Formula**:
 $$
  \sigma(x) = \frac{1}{1 + e^{-x}}
  $$
  
- **Behavior**: The sigmoid function squishes any real number into the range between 0 and 1. It's commonly used for binary classification tasks because it can output values representing probabilities.

---

### 2. **Hyperbolic Tangent (tanh)**

- **Formula**:
  $$
  \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
  $$
  
- **Behavior**: The tanh function squishes the input into a range between -1 and 1. It is often preferred over the sigmoid function because it is zero-centered, which helps mitigate issues with gradients in neural networks.

---

### 3. **Rectified Linear Unit (ReLU)**

- **Formula**:
  $$
  \text{ReLU}(x) = \max(0, x)
  $$
  
- **Behavior**: The ReLU function outputs the input value directly if it's positive, and 0 otherwise. It is widely used in deep learning because of its simplicity and efficiency in handling positive and negative values.

---

### 4. **Leaky ReLU**

- **Formula**:
  $$
  \text{Leaky ReLU}(x) = \max(\alpha x, x)
  $$
  Where $\alpha$ is a small positive constant, typically $\alpha = 0.01$.

- **Behavior**: Like ReLU, but instead of outputting 0 for negative inputs, it outputs a small fraction ($\alpha$) of the input. This helps prevent the "dying ReLU" problem, where neurons get stuck and stop updating.

---

### 5. **Softmax Function**

- **Formula**:
  $$
  \text{softmax}(x)_i = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
  $$
  Where $x_i$ is the $i$th element in the vector $x$, and the denominator is the sum of exponentials over all elements in the vector $x$.

- **Behavior**: The softmax function converts a vector of real numbers into a probability distribution. Each element in the output will be between 0 and 1, and the sum of all elements will equal 1. It is mainly used in the output layer for multi-class classification tasks.
