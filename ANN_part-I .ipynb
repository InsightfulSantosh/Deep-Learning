### **Artificial Neural Networks (ANN) - Study Notes**

**Definition:**
An **Artificial Neural Network (ANN)** is a computational model inspired by the way biological neural networks in the human brain function. ANNs consist of interconnected nodes, known as neurons, that process and transmit information in a layered structure. They are used to recognize patterns, classify data, make predictions, and perform various complex tasks like image recognition, speech processing, and language translation.

---

### **Key Components of an ANN:**

####1. **Neurons (Nodes):**

image.png

   - The basic units of an ANN, analogous to neurons in the brain.
   - Each neuron takes input, processes it, and produces an output that serves as input for the next layer.
   
####2. **Weights:**


image.png

   - Each connection between neurons has an associated weight that represents the strength of that connection.
   - Weights are adjusted during training to improve the accuracy of the network.


####3. **Bias:**

image.png

   - A bias value is added to the weighted sum of inputs to adjust the output independently of the inputs.
   - Bias helps the network to better fit the data.

###4. **Activation Function:**


image.png

   - Determines whether a neuron should be activated or not based on the input.
   - Common activation functions include:
     - **Sigmoid:** Output is in the range (0, 1).
     - **ReLU (Rectified Linear Unit):** Outputs the input directly if positive; otherwise, it outputs zero.
     - **Tanh:** Output is in the range (-1, 1).

---

### **Structure of an ANN:**

ANNs consist of three types of layers:

1. **Input Layer:**
   - The first layer, where the raw data is fed into the network.
   - Each node in the input layer represents a feature or input variable from the dataset.

2. **Hidden Layer(s):**
   - Layers between the input and output layers.
   - Perform computations on the inputs and learn complex features.
   - A deep network may have multiple hidden layers, making it a "deep" neural network.

3. **Output Layer:**
   - The final layer that produces the network's prediction or classification.
   - The number of nodes in the output layer corresponds to the number of classes (in classification) or the target value (in regression).

---

### **Types of Artificial Neural Networks:**

1. **Feedforward Neural Network (FNN):**
   - The simplest form of ANN where the connections between the nodes do not form cycles.
   - Information moves in one direction, from input to output.
   - Suitable for tasks like image classification and speech recognition.

2. **Convolutional Neural Network (CNN):**
   - Primarily used for image and video recognition.
   - Uses convolutional layers to capture spatial and temporal dependencies.
   - Consists of three main types of layers: convolutional, pooling, and fully connected.

3. **Recurrent Neural Network (RNN):**
   - Designed for sequential data like time series, speech, or text.
   - Unlike FNNs, RNNs have connections that form cycles, allowing them to maintain a memory of previous inputs.
   - A popular variant is **Long Short-Term Memory (LSTM)**, which helps overcome short-term memory problems.

4. **Radial Basis Function Neural Network (RBFNN):**
   - Uses radial basis functions as activation functions.
   - Mainly used for pattern classification and function approximation.

5. **Autoencoders:**
   - Used for unsupervised learning.
   - Designed to learn efficient representations of input data, typically for dimensionality reduction.
   - Consists of an encoder (to compress the input) and a decoder (to reconstruct the input).

6. **Generative Adversarial Networks (GANs):**
   - Composed of two networks: a generator and a discriminator.
   - The generator creates data, and the discriminator tries to differentiate between real and generated data.
   - GANs are widely used for image generation and deepfake creation.

### **Perceptron:**

The **Perceptron** is the simplest form of a neural network model, consisting of a single neuron. It was the first model of an artificial neuron and forms the foundation for more complex networks.

#### **Perceptron Model:**
1. **Inputs (x1, x2, , xn):**
   - The inputs represent the features of the data that we want to classify.

2. **Weights (w1, w2, , wn):**
   - Each input is multiplied by a corresponding weight, which is learned during training.

3. **Summation:**
   - The weighted inputs are summed up and passed through an activation function.
   $$
   Z = \sum (w_i * x_i) + b
   $$
   Where $(b)$ is the bias term.

4. **Activation Function:**
   - The result of the summation is passed through an activation function to get the output.
   - For binary classification, a step function or sigmoid function is often used:
   $$
   f(Z) = \left\{ \begin{array}{ll}
   1 & \text{if } Z \geq \text{threshold}\\
   0 & \text{if } Z < \text{threshold}
   \end{array} \right.
   $$

#### **Learning in a Perceptron:**
- The perceptron is trained using a learning algorithm that adjusts weights based on the error in its predictions.
- The weights are updated according to the perceptron learning rule:
  $$
  w_{i} = w_{i} + \Delta w_{i}
  $$
  $$
  \Delta w_{i} = \eta (y - \hat{y}) x_{i}
  $$
  Where:
  - $( \eta )$ is the learning rate.
  - $( y )$ is the actual output.
  - $( \hat{y} )$ is the predicted output.

---

### **Multi-Layer Perceptron (MLP):**
A **Multi-Layer Perceptron** is a feedforward neural network with one or more hidden layers between the input and output layers. It addresses the limitation of a single-layer perceptron, which can only handle linearly separable data.

- **Key Characteristics of MLP:**
  - Consists of at least one hidden layer.
  - Uses nonlinear activation functions like ReLU, sigmoid, or tanh.
  - Trained using the **backpropagation** algorithm, which calculates the error at the output and propagates it back to update weights.

---

### **Backpropagation in ANN:**
Backpropagation is the core algorithm used to train ANNs. It involves the following steps:
1. **Forward Pass:** Compute the output by passing the input through the network.
2. **Loss Calculation:** Calculate the difference between the predicted output and the actual output (error).
3. **Backward Pass:** Propagate the error backward through the network and update the weights using gradient descent to minimize the error.

### **Single Layer Perceptron:**

A **Single Layer Perceptron (SLP)** is the most basic type of artificial neural network. It consists of only two layers: the **input layer** and the **output layer**. The neurons in the input layer are connected directly to the neurons in the output layer, with no intermediate (hidden) layers.

#### **Structure of a Single Layer Perceptron:**
1. **Input Layer:**
   - Contains neurons (nodes) representing input features.
   - Each input is connected to the output neuron(s) via weighted connections.

2. **Output Layer:**
   - Contains a single or multiple neurons, depending on the problem (binary or multi-class classification).
   - The output is a linear combination of the inputs and their corresponding weights, followed by an activation function.

#### **Working of a Single Layer Perceptron:**
- **Input:** Each input value is multiplied by a weight.
- **Summation:** The weighted inputs are summed, and a bias is added.
  $$
  Z = \sum (w_i * x_i) + b
  $$
- **Activation Function:** The sum is passed through an activation function, which decides the output. Common choices include:
- **Step function** for binary classification:
$$
f(Z) = \left\{ \begin{array}{ll}
1 & \text{if } Z \geq \text{threshold} \\
0 & \text{if } Z < \text{threshold}
\end{array} \right.
$$
  - **Sigmoid function** for outputs between 0 and 1.

#### **Key Characteristics:**
- **Binary Classification:** A single-layer perceptron is mainly used for binary classification tasks.
- **Linearly Separable Data:** SLPs can only solve problems where the data is **linearly separable**, meaning the decision boundary is a straight line or hyperplane.
  
#### **Limitations:**
- A single-layer perceptron cannot solve non-linear problems like XOR because it lacks the ability to form complex decision boundaries.
  
---

### **Multi-Layer Perceptron (MLP):**

A **Multi-Layer Perceptron (MLP)** is a more complex type of neural network that contains one or more **hidden layers** between the input and output layers. It overcomes the limitations of the single-layer perceptron by introducing non-linear layers, making it capable of solving non-linearly separable problems.

#### **Structure of a Multi-Layer Perceptron:**
1. **Input Layer:**
   - Same as in SLP, the input layer provides data to the hidden layer(s).

2. **Hidden Layer(s):**
   - One or more intermediate layers where complex computations occur.
   - Each hidden layer neuron receives inputs from all neurons in the previous layer, computes a weighted sum, and applies a non-linear activation function (e.g., ReLU, sigmoid, or tanh).

3. **Output Layer:**
   - Generates the final prediction based on the processed inputs from the last hidden layer.

#### **Working of a Multi-Layer Perceptron:**
- **Forward Propagation:**
  - Inputs are passed through each layer, and at each neuron, a weighted sum is computed, followed by applying an activation function.
  
- **Activation Functions:**
  - **ReLU (Rectified Linear Unit):** Commonly used in hidden layers, it outputs the input if positive, otherwise outputs zero:
    $$
    f(x) = \text{max}(0, x)
    $$
  - **Sigmoid:** Often used in output layers for binary classification:
    $$
    f(x) = \frac{1}{1 + e^{-x}}
    $$

- **Backpropagation:**
  - The error at the output is calculated using a loss function, and then this error is propagated backward through the network to update the weights using an optimization algorithm like **gradient descent**.

#### **Key Characteristics:**
- **Non-linear Problems:** MLP can model more complex, non-linearly separable data.
- **Hidden Layers:** Hidden layers allow MLPs to capture intricate patterns in the data, improving their predictive ability.
- **Deep Learning:** If an MLP has many hidden layers, it becomes a **deep neural network** (DNN).

#### **Advantages:**
- **Solves XOR Problem:** Unlike single-layer perceptrons, MLPs can solve non-linear problems like the XOR problem.
- **Versatility:** MLPs are highly versatile and can handle a wide variety of tasks, including regression, classification, and function approximation.

#### **Limitations:**
- **Training Complexity:** MLPs require more computational resources and training time compared to single-layer perceptrons.
- **Overfitting:** Without proper regularization, MLPs can overfit to the training data, especially with a large number of hidden layers or neurons.

In conclusion, while a **Single Layer Perceptron** is limited to solving linearly separable problems, a **Multi-Layer Perceptron** can handle more complex, non-linear tasks due to the inclusion of hidden layers and non-linear activation functions. The MLP's ability to learn from errors through backpropagation makes it a powerful tool in machine learning.

## **Structure of Artificial Neural Networks (ANNs)**

The structure of an Artificial Neural Network (ANN) is designed to mimic the way biological neural networks process information. Heres a detailed description of the components that make up an ANN:

--------

#### 1. **Neurons (Nodes)**
- **Neurons** are the fundamental units of an ANN. Each neuron receives input, processes it, and passes the output to the next layer.
- **Types of Neurons**:
  - **Input Neurons**: Receive external data.
  - **Hidden Neurons**: Process data within hidden layers.
  - **Output Neurons**: Provide the final result.

------

#### 2. **Layers**
ANNs are composed of multiple layers:

- **Input Layer**:
  - **Purpose**: Receives the raw data input.
  - **Structure**: Consists of neurons that correspond to each feature in the input data.
  - **Example**: For image data, each pixel would correspond to a neuron in the input layer.

- **Hidden Layers**:
  - **Purpose**: Perform computations and extract features from the input data.
  - **Structure**: Can have one or more hidden layers. Each layer consists of neurons that process inputs from the previous layer.
  - **Activation Functions**: Apply functions like ReLU, Sigmoid, or Tanh to introduce non-linearity and help the network learn complex patterns.
  - **Example**: In a neural network for image recognition, hidden layers might detect edges, textures, or shapes.

- **Output Layer**:
  - **Purpose**: Provides the final output based on the computations performed in the hidden layers.
  - **Structure**: Contains neurons corresponding to the possible outcomes or classes. For binary classification, there might be a single neuron; for multi-class classification, there will be one neuron per class.
  - **Activation Functions**: Often uses functions like Softmax (for multi-class classification) or Sigmoid (for binary classification) to produce probabilities or scores.

-------

#### 3. **Weights and Biases**
- **Weights**:
  - **Purpose**: Adjust the strength of the connection between neurons.
  - **Function**: Each connection between neurons has an associated weight that determines the influence of one neurons output on another neurons input.
  - **Training**: Weights are adjusted during training to minimize the error in predictions.

- **Biases**:
  - **Purpose**: Allow the activation function to be shifted to the left or right, which helps in better fitting the model.
  - **Function**: Added to the weighted sum of inputs before applying the activation function.

-----------

#### 4. **Activation Functions**
- **Purpose**: Introduce non-linearity into the model, enabling it to learn complex patterns.
- **Types**:
- **ReLU (Rectified Linear Unit)**:
$$
f(x) = \text{max}(0, x)
$$
- Commonly used in hidden layers.
- **Sigmoid**:
$$
f(x) = \frac{1}{1 + e^{-x}}
$$
- Used in binary classification tasks.
- **Tanh (Hyperbolic Tangent)**:
$$
f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$
- Provides outputs in the range [-1, 1].
- **Softmax**:
$$
f(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
$$
- Used in multi-class classification to provide probabilities for each class.

-------

#### 5. **Forward Propagation**
- **Process**: Input data is passed through the network from the input layer to the output layer.
- **Steps**:
  1. **Weighted Sum**: Compute the weighted sum of inputs plus biases.
  2. **Activation**: Apply the activation function to the weighted sum.

-----------

#### 6. **Backpropagation**
- **Purpose**: Update weights and biases based on the error between predicted and actual output.
- **Process**:
  1. **Error Calculation**: Compute the difference between the predicted output and the actual output using a loss function.
  2. **Gradient Descent**: Use optimization algorithms like gradient descent to adjust weights and biases to minimize the error.
  3. **Propagation**: Propagate the error backward through the network, adjusting weights and biases using the gradients computed.

------

### **Summary**

The structure of an ANN involves:
- **Neurons** arranged in **layers** (input, hidden, output).
- **Weights** and **biases** associated with connections between neurons.
- **Activation functions** to introduce non-linearity.
- **Forward propagation** to compute outputs.
- **Backpropagation** to optimize the network through learning.

Each of these components works together to enable ANNs to learn from data and make accurate predictions or classifications.

### **1. Input Layer**

**Purpose:** The input layer is where the data enters the neural network. It is responsible for receiving and passing the raw data to the subsequent layers.

**Characteristics:**
- **Nodes:** Each node (neuron) in the input layer represents a feature of the input data. For example, in an image classification task, each pixel value might correspond to a node.
- **No Computation:** The input layer does not perform any computations. It simply forwards the input data to the next layer.
- **Feature Representation:** The number of neurons in the input layer corresponds to the number of features in the input data.

**Example:**
For an image of size 28x28 pixels, the input layer will have 784 neurons (one for each pixel).

---

### **2. Hidden Layers**

**Purpose:** Hidden layers perform computations and feature transformations on the data received from the input layer. They enable the network to learn complex patterns and representations.

**Characteristics:**
- **Nodes:** Neurons in hidden layers are interconnected, and each neuron computes a weighted sum of its inputs, applies an activation function, and passes the result to the next layer.
- **Activation Functions:** Common activation functions used in hidden layers include:
  - **ReLU (Rectified Linear Unit):**
$$
f(x) = \text{max}(0, x)
$$
Helps introduce non-linearity and is widely used due to its simplicity and effectiveness.
- **Sigmoid:**
$$
f(x) = \frac{1}{1 + e^{-x}}
$$
Maps the output to a range between 0 and 1, often used in binary classification.
- **Tanh (Hyperbolic Tangent):**
$$
f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$
Maps the output to a range between -1 and 1.
- **Depth:** ANNs can have one or multiple hidden layers. Networks with more than one hidden layer are called **Deep Neural Networks (DNNs)**.

**Example:**
In a DNN for image classification, there might be multiple hidden layers, each extracting different levels of features from the image (e.g., edges, textures, shapes).

---

### **3. Output Layer**

**Purpose:** The output layer provides the final result of the network's computation, such as predictions or classifications.

**Characteristics:**
- **Nodes:** The number of neurons in the output layer depends on the type of task:
  - **Classification:** For binary classification, there is typically one output neuron with a sigmoid activation function. For multi-class classification, there are as many neurons as classes, often using the softmax activation function.
  - **Regression:** For regression tasks, the output layer usually has one neuron with a linear activation function.
- **Activation Functions:** Common activation functions include:
  - **Softmax:**
$$
f(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
$$
Converts outputs to a probability distribution across multiple classes.
- **Linear Activation:**
$$
f(x) = x
$$
Used for regression tasks where output can be any real value.

**Example:**
In a digit classification task (0-9), the output layer will have 10 neurons, each representing a digit class, with a softmax activation function.

---

### **4. Bias**

**Purpose:** Biases are added to the weighted sum of inputs to help the model fit the data better. They allow the activation function to be shifted to the left or right, providing more flexibility.

**Characteristics:**
- **Addition:** Bias is added to the weighted sum before applying the activation function.
- **Parameter:** Each neuron typically has its own bias parameter.

**Example:**
In a neuron with inputs $( x_1, x_2 )$ and weights $( w_1, w_2 )$, the output before activation is:
$$
Z = w_1 \cdot x_1 + w_2 \cdot x_2 + b
$$
where $( b )$ is the bias.

---

### **5. Dropout Layer**

**Purpose:** The dropout layer is used as a regularization technique to prevent overfitting by randomly dropping a fraction of neurons during training.

**Characteristics:**
- **Random Dropping:** During each training iteration, a percentage of neurons are randomly set to zero.
- **Training vs. Testing:** Dropout is applied during training but not during testing.

**Example:**
A dropout rate of 0.5 means that each neuron has a 50% chance of being dropped out during each training iteration.

---

### **6. Normalization Layers**

**Purpose:** Normalization layers help stabilize and accelerate the training process by normalizing the inputs to each layer.

**Characteristics:**
- **Batch Normalization:** Normalizes activations across the batch for each layer.
- **Layer Normalization:** Normalizes activations across each layers features independently.

**Example:**
Batch normalization adjusts the output to have a mean of zero and a standard deviation of one, which helps in faster and more stable training.

- **Input Layer**: Receives data and passes it to hidden layers.
- **Hidden Layers**: Perform computations and learn features from the data. Can have multiple layers.
- **Output Layer**: Produces final predictions or classifications.
- **Bias**: Provides flexibility by allowing shifting of the activation function.
- **Dropout Layer**: Regularization technique to prevent overfitting.
- **Normalization Layers**: Stabilize and accelerate training by normalizing activations.

# **Forward Propagation** and **Backward Propagation** in Artificial Neural Networks (ANNs)

### **Forward Propagation**

**Forward Propagation** is the process of passing input data through the neural network to obtain the output. It involves calculating the activations of neurons in each layer.

#### **Steps in Forward Propagation:**

1. **Input Layer**:
   - **Input Data**: $( x = [x_1, x_2, ..., x_n] )$ (for example, pixel values of an image)
   
2. **Hidden Layer(s)**:
   - Each neuron in the hidden layer calculates a weighted sum of inputs plus a bias term, then applies an activation function.
   - For a neuron $( j )$ in the hidden layer:
$$
z_j = \sum_{i=1}^{n} w_{ji} x_i + b_j
$$
where $( w_{ji} )$ is the weight from input neuron $( i )$ to hidden neuron $( j )$, and $( b_j )$ is the bias for neuron $( j )$.
   
   - The activation function $( f )$ (like ReLU, Sigmoid, or Tanh) is applied:
     $$
     a_j = f(z_j)
     $$
     where $( a_j )$ is the output of neuron $( j )$ in the hidden layer.

3. **Output Layer**:
   - Similar to hidden layers, neurons in the output layer compute:
     $$
     z_k = \sum_{j=1}^{m} w_{kj} a_j + b_k
     $$
     where $( m )$ is the number of neurons in the previous layer.
   
   - Apply the activation function to get the final output \( \hat{y} \):
     $$
     \hat{y} = f(z_k)
     $$

#### **Example:**
Consider a simple neural network with one hidden layer.

- **Input Layer**: $( x = [2, 3] )$
- **Hidden Layer**: 2 neurons
  - Weights: $( W = \begin{bmatrix} 0.5 & 0.2 \\ 0.1 & 0.4 \end{bmatrix} )$
  - Biases: $( b = [0.1, 0.2] )$
  - Activation Function: Sigmoid $( f(z) = \frac{1}{1 + e^{-z}} )$

1. **Hidden Layer Calculations:**
   $$
   z_1 = 0.5 \times 2 + 0.2 \times 3 + 0.1 = 1.4
   $$
   $$
   z_2 = 0.1 \times 2 + 0.4 \times 3 + 0.2 = 1.7
   $$
   $$
   a_1 = \frac{1}{1 + e^{-1.4}} \approx 0.802
   $$
   $$
   a_2 = \frac{1}{1 + e^{-1.7}} \approx 0.845
   $$

2. **Output Layer Calculation:**
   - Weights and biases are specific to the network design.

---------------
### **Backward Propagation**

**Backward Propagation** is used to update the weights and biases of the network to minimize the loss function. It involves calculating the gradient of the loss function with respect to each weight and bias.

#### **Steps in Backward Propagation:**

1. **Compute Loss:**
   - The loss function measures the difference between the predicted output $( \hat{y} )$ and the true output $( y )$.
   - Common loss functions: Mean Squared Error (MSE) for regression, Cross-Entropy Loss for classification.

2. **Calculate Gradient of Loss with Respect to Output:**
- For a loss function $( L )$, compute:
$$
\frac{\partial L}{\partial \hat{y}}
$$

3. **Backpropagate the Error:**
   - Compute the gradient of the loss with respect to each neurons activation and weights using the chain rule.
   - For each neuron $( k )$ in the output layer:
$$
\delta_k = \frac{\partial L}{\partial \hat{y}} \cdot f'(z_k)
$$
where $(f'(z_k))$ is the derivative of the activation function.

   - For hidden layers, propagate the error backward:
     $$
     \delta_j = \left( \sum_{k=1}^{m} \delta_k \cdot w_{kj} \right) \cdot f'(z_j)
     $$

4. **Update Weights and Biases:**
   - Use the calculated gradients to update the weights and biases:
     $$
     w_{ji} \leftarrow w_{ji} - \eta \cdot \frac{\partial L}{\partial w_{ji}}
     $$
     $$
     b_j \leftarrow b_j - \eta \cdot \frac{\partial L}{\partial b_j}
     $$
     where $( \eta \)$ is the learning rate.

#### **Example:**

1. **Calculate Loss:**
   - Assume the true output $( y = 1 )$ and predicted output $( \hat{y} = 0.8 )$.
   - Using MSE:
     $$
     L = \frac{1}{2} (\hat{y} - y)^2 = \frac{1}{2} (0.8 - 1)^2 = 0.02
     $$

2. **Compute Gradient of Loss:**
   - For output layer:
     $$
     \frac{\partial L}{\partial \hat{y}} = \hat{y} - y = 0.8 - 1 = -0.2
     $$
   - For the activation function:
     $$
     f'(z_k) = \hat{y} \cdot (1 - \hat{y})
     $$

3. **Calculate Gradients and Update Weights:**
   - For output layer weight \( w_{kj} \):
     $$
     \delta_k = -0.2 \cdot 0.8 \cdot (1 - 0.8) = -0.032
     $$
     Update weight:
     $$
     w_{kj} \leftarrow w_{kj} - \eta \cdot (-0.032)
     $$

-------------
### **Summary**

- **Forward Propagation** calculates the output of the network by propagating the inputs through the layers using weights, biases, and activation functions.
- **Backward Propagation** adjusts weights and biases by calculating the gradient of the loss function with respect to these parameters, using the chain rule and gradient descent to minimize the loss.

Both processes are fundamental to training neural networks, enabling them to learn from data and make accurate predictions.

### **Activation Functions and Their Derivatives**
----------
### **1. Sigmoid Activation Function**

- **Function:**
  $$
  f(z) = \frac{1}{1 + e^{-z}}
  $$
  The sigmoid function maps any real-valued number into the (0, 1) interval, making it suitable for binary classification problems.

- **Derivative:**
  To find the derivative, we use the fact that:
  $$
  f(z) = \frac{1}{1 + e^{-z}}
  $$
  Let $( f(z) = \sigma(z) )$. Then:
  $$
  \sigma(z) = \frac{1}{1 + e^{-z}}
  $$
  The derivative of \( \sigma(z) \) can be computed as:
  $$
  \frac{d\sigma(z)}{dz} = \sigma(z) \cdot (1 - \sigma(z))
  $$
  This follows from:
  $$
  \frac{d\sigma(z)}{dz} = \frac{e^{-z}}{(1 + e^{-z})^2}
  $$
  and since $( \sigma(z) = \frac{1}{1 + e^{-z}} )$, substituting $( \sigma(z) )$ in gives:
  $$
  \frac{d\sigma(z)}{dz} = \sigma(z) \cdot (1 - \sigma(z))
  $$

-----------

### **2. ReLU (Rectified Linear Unit)**

- **Function:**
  $$
  f(z) = \max(0, z)
  $$
  ReLU is popular due to its simplicity and effectiveness in mitigating the vanishing gradient problem.

- **Derivative:**
  The derivative of ReLU is:
  $$
  \frac{d}{dz} \max(0, z) =
  \begin{cases}
  1 & \text{if } z > 0 \\
  0 & \text{if } z \leq 0
  \end{cases}
  $$
  This means the gradient is 1 when the input is positive and 0 otherwise.

----------

### **3. Tanh (Hyperbolic Tangent)**

- **Function:**
  $$
  f(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}
  $$
  The tanh function maps values to the range (-1, 1), and is zero-centered, which helps in training deep networks.

- **Derivative:**
  The derivative of the tanh function is:
  $$
  \frac{d}{dz} \left( \frac{e^z - e^{-z}}{e^z + e^{-z}} \right) = 1 - \tanh^2(z)
  $$
  This comes from the property:
  $$
  \frac{d \tanh(z)}{dz} = 1 - \tanh^2(z)
  $$

-------

### **4. Softmax Function**

- **Function:**
  For a vector $( \mathbf{z} )$ with components $( z_i )$, the softmax function is defined as:
  $$
  f(z_i) = \frac{e^{z_i}}{\sum_{j} e^{z_j}}
  $$
  Softmax is used in the output layer of classification problems where we want to predict the probabilities of each class.

- **Derivative:**
  The derivative of the softmax function is more complex and involves the Jacobian matrix. For $( f(z_i) )$, the partial derivative with respect to $( z_j )$ is:
  $$
  \frac{\partial f(z_i)}{\partial z_j} = f(z_i) \cdot (\delta_{ij} - f(z_j))
  $$
  where $( \delta_{ij} )$ is the Kronecker delta, which is 1 if $( i = j )$ and 0 otherwise.

---------

### **Summary**

- **Sigmoid**: $( f(z) = \frac{1}{1 + e^{-z}} )$, derivative $( f'(z) = f(z) \cdot (1 - f(z)) )$
- **ReLU**: $( f(z) = \max(0, z) )$, derivative $( f'(z) = 1 \text{ if } z > 0, 0 \text{ otherwise} )$
- **Tanh**: $( f(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}} )$, derivative $( f'(z) = 1 - \tanh^2(z) )$
- **Softmax**: $( f(z_i) = \frac{e^{z_i}}{\sum_{j} e^{z_j}} )$, derivative involves the Jacobian matrix $( \frac{\partial f(z_i)}{\partial z_j} = f(z_i) \cdot (\delta_{ij} - f(z_j)) )$

These activation functions and their derivatives are essential in adjusting the weights and biases of neural networks during training, which allows the network to learn and make accurate predictions.

-----------

### **Backpropagation with Activation Function Derivatives**

During backpropagation, you use the derivative of the activation function to calculate how much each weight in the network contributed to the error. For the output layer, if you have the gradient of the loss function with respect to the output $( \delta_k )$, you calculate:

$$
\delta_k = \frac{\partial L}{\partial \hat{y}} \cdot f'(z_k)
$$

where:

- $( \frac{\partial L}{\partial \hat{y}} )$ is the gradient of the loss function with respect to the predicted output.
- $( f'(z_k) )$ is the derivative of the activation function with respect to the input $( z_k )$ to the neuron.

This helps in updating the weights by determining how much change in weights will reduce the loss. The gradient descent step uses this information to adjust weights and biases in the direction that reduces the error.

# **Example for Regression Problem**:
Based on the setup , we're designing a neural network with the following architecture and components:

1. **Input Layer**: 2 neurons
2. **Hidden Layer**: 2 neurons with sigmoid activation
3. **Output Layer**: 1 neuron with linear activation
4. **Loss Function**: Mean Squared Error (MSE)
5. **Learning Rate**: 0.1

Here's a breakdown of how this network will function:

### **Architecture**:

- **Input Layer**: Takes in 2 input features.
- **Hidden Layer**: The 2 neurons in this layer apply the sigmoid activation function. Sigmoid activation is defined as $(\sigma(x) = \frac{1}{1 + e^{-x}})$, which maps inputs to a range between 0 and 1.
- **Output Layer**: The single neuron in the output layer has a linear activation function, meaning it will output a value that is a linear combination of the inputs from the hidden layer.

### **Forward Propagation**:

1. **Input Layer to Hidden Layer**:
   - Compute weighted sum for each neuron in the hidden layer.
   - Apply the sigmoid activation function to these sums to get the hidden layer outputs.

2. **Hidden Layer to Output Layer**:
   - Compute the weighted sum of the hidden layer outputs for the output neuron.
   - Apply the linear activation function (which is just the identity function) to get the final output.

### **Loss Calculation**:

- **Loss Function**: Mean Squared Error (MSE) is used to measure the difference between the predicted output and the actual target value. It's calculated as:
  $$
  \text{MSE} = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2
  $$
  where $(y_i)$ is the true value and $(\hat{y}_i)$ is the predicted value, and $(N)$ is the number of samples.

### **Backpropagation and Weight Update**:

- Compute the gradient of the MSE loss with respect to each weight using backpropagation.
- Update the weights using the gradient descent algorithm with the learning rate of 0.1.

### **Gradient Descent Update Rule**:

For each weight $(w)$, the update rule is:
$$
w = w - \text{learning rate} \times \frac{\partial \text{MSE}}{\partial w}
$$

### **Training Process**:

1. **Initialize weights**: Randomly initialize weights for the network.
2. **Forward Propagation**: Pass inputs through the network to get predictions.
3. **Compute Loss**: Calculate the MSE loss between predictions and actual values.
4. **Backpropagation**: Compute gradients and update weights.
5. **Repeat**: Continue the process for multiple epochs until the loss converges or achieves a satisfactory level.

### **Setup**:
- **Input Layer**: 2 neurons
- **Hidden Layer**: 2 neurons with sigmoid activation
- **Output Layer**: 1 neuron with **linear activation**
- **Loss Function**: Mean Squared Error (MSE)
- **Learning Rate**: 0.1

### **Weights and Biases Initialization** (unchanged):
- **Input to Hidden Weights $( W_1 )$**:
$$
W_1 = \begin{bmatrix} 0.15 & 0.25 \\ 0.20 & 0.30 \end{bmatrix}
$$
- **Hidden Layer Bias $( b_1 )$**:
$$
b_1 = \begin{bmatrix} 0.35 \\ 0.35 \end{bmatrix}
$$
- **Hidden to Output Weights $( W_2 )$**:
$$
W_2 = \begin{bmatrix} 0.40 & 0.45 \end{bmatrix}
$$
- **Output Layer Bias $( b_2 )$**:
$$
b_2 = \begin{bmatrix} 0.60 \end{bmatrix}
$$

### **Input and Target Output**:
- **Input $( X )$**:
$$
X = \begin{bmatrix} 0.05 \\ 0.10 \end{bmatrix}
$$
- **Target Output $( y )$**:
$$
y = 0.01
$$

---

## **Forward Propagation for Regression**

### **Step 1: Hidden Layer Computation** (unchanged)
Calculate the input to the hidden layer:

$$
Z_1 = W_1 \cdot X + b_1
$$
$$
Z_1 = \begin{bmatrix} 0.15 & 0.25 \\ 0.20 & 0.30 \end{bmatrix} \cdot \begin{bmatrix} 0.05 \\ 0.10 \end{bmatrix} + \begin{bmatrix} 0.35 \\ 0.35 \end{bmatrix}
$$
$$
Z_1 = \begin{bmatrix} 0.3775 \\ 0.3925 \end{bmatrix}
$$

Apply the sigmoid activation function:

$$
A_1 = \sigma(Z_1) = \frac{1}{1 + e^{-Z_1}} = \begin{bmatrix} 0.5933 \\ 0.5969 \end{bmatrix}
$$

### **Step 2: Output Layer Computation** (modified for regression)

Since we are using a linear activation for the output layer (i.e., no activation function), the output is simply:

$$
Z_2 = W_2 \cdot A_1 + b_2
$$
$$
Z_2 = \begin{bmatrix} 0.40 & 0.45 \end{bmatrix} \cdot \begin{bmatrix} 0.5933 \\ 0.5969 \end{bmatrix} + 0.60
$$
$$
Z_2 = (0.40 \cdot 0.5933) + (0.45 \cdot 0.5969) + 0.60 = 1.1059
$$

Here, $( Z_2 )$ is the predicted output $( \hat{y} )$.

$$
\hat{y} = Z_2 = 1.1059
$$

### **Step 3: Compute the Loss**

We will use the **Mean Squared Error (MSE)** loss function to calculate the loss:

$$
L = \frac{1}{2}(y - \hat{y})^2 = \frac{1}{2}(0.01 - 1.1059)^2 = \frac{1}{2}(-1.0959)^2 = 0.6004
$$

---

## **Backward Propagation for Regression**

### **Step 4: Compute the Gradient for the Output Layer**

Since the output is linear, the derivative of the output with respect to $( Z_2 )$ is simply 1.

The gradient of the loss with respect to $( Z_2 )$ (i.e., $( \hat{y} )$) is:

$$
\frac{\partial L}{\partial Z_2} = \hat{y} - y = 1.1059 - 0.01 = 1.0959
$$

### **Step 5: Compute the Gradients for $( W_2 )$ and $( b_2 )$**

The gradient of the loss with respect to the weights $( W_2 )$ connecting the hidden layer to the output is:

$$
\frac{\partial L}{\partial W_2} = \frac{\partial L}{\partial Z_2} \cdot A_1
$$
$$
\frac{\partial L}{\partial W_2} = 1.0959 \cdot \begin{bmatrix} 0.5933 \\ 0.5969 \end{bmatrix} = \begin{bmatrix} 0.6503 \\ 0.6536 \end{bmatrix}
$$

The gradient of the loss with respect to the output bias $( b_2 )$ is:

$$
\frac{\partial L}{\partial b_2} = \frac{\partial L}{\partial Z_2} = 1.0959
$$

### **Step 6: Backpropagate the Error to the Hidden Layer**

To backpropagate the error to the hidden layer, we calculate the gradient of the loss with respect to the activations $( A_1 )$:

$$
\frac{\partial L}{\partial A_1} = \frac{\partial L}{\partial Z_2} \cdot W_2
$$
$$
\frac{\partial L}{\partial A_1} = 1.0959 \cdot \begin{bmatrix} 0.40 \\ 0.45 \end{bmatrix} = \begin{bmatrix} 0.4384 \\ 0.4932 \end{bmatrix}
$$

Now, compute the gradient with respect to $( Z_1 )$ using the derivative of the sigmoid activation function:

$$
\sigma'(Z_1) = A_1(1 - A_1) = \begin{bmatrix} 0.5933 \cdot (1 - 0.5933) \\ 0.5969 \cdot (1 - 0.5969) \end{bmatrix}
$$
$$
\sigma'(Z_1) = \begin{bmatrix} 0.2413 \\ 0.2406 \end{bmatrix}
$$

Thus, the gradient of the loss with respect to $( Z_1 )$ is:

$$
\frac{\partial L}{\partial Z_1} = \frac{\partial L}{\partial A_1} \cdot \sigma'(Z_1)
$$
$$
\frac{\partial L}{\partial Z_1} = \begin{bmatrix} 0.4384 \\ 0.4932 \end{bmatrix} \cdot \begin{bmatrix} 0.2413 \\ 0.2406 \end{bmatrix} = \begin{bmatrix} 0.1057 \\ 0.1187 \end{bmatrix}
$$

### **Step 7: Compute the Gradients for $( W_1 )$ and $( b_1 )$**

The gradient for the input-to-hidden weights $( W_1 )$ is:

$$
\frac{\partial L}{\partial W_1} = \frac{\partial L}{\partial Z_1} \cdot X^T
$$
$$
\frac{\partial L}{\partial W_1} = \begin{bmatrix} 0.1057 \\ 0.1187 \end{bmatrix} \cdot \begin{bmatrix} 0.05 & 0.10 \end{bmatrix}
$$
$$
\frac{\partial L}{\partial W_1} = \begin{bmatrix} 0.00528 & 0.01057 \\ 0.00593 & 0.01187 \end{bmatrix}
$$
The gradient for $( b_1 )$ is:

$$
\frac{\partial L}{\partial b_1} = \frac{\partial L}{\partial Z_1} = \begin{bmatrix} 0.1057 \\ 0.1187 \end{bmatrix}
$$

---

## **Step 8: Update the Weights and Biases**

Finally, update the weights and biases using gradient descent with a learning rate of $( \alpha = 0.1 )$:

- $$( W_2 = W_2 - \alpha \cdot \frac{\partial L}{\partial W_2} = \begin{bmatrix}
0.40 \\ 0.45 \end{bmatrix} - 0.1 \cdot \begin{bmatrix} 0.6503 \\ 0.6536 \end{bmatrix} = \begin{bmatrix} 0.33497 \\ 0.38464 \end{bmatrix} )$$
- $$( b_2 = b_2 - \alpha \cdot \frac{\partial L}{\partial b_2} = 0.60 - 0.1 \cdot 1.0959 = 0.49041 )$$
- $$( W_1 = W_1 - \alpha \cdot \frac{\partial L}{\partial W_1} = \begin{bmatrix} 0.15 & 0.25 \\ 0.20 & 0.30 \end{bmatrix} - 0.1 \cdot \begin{bmatrix} 0.00528 & 0.01057 \\ 0.00593 & 0.01187 \end{bmatrix} = \begin{bmatrix} 0.14947 & 0.24894 \\ 0.19941 & 0.29881 \end{bmatrix})$$
- $$( b_1 = b_1 - \alpha \cdot \frac{\partial L}{\partial b_1} = \begin{bmatrix} 0.35 \\ 0.35 \end{bmatrix} - 0.1 \cdot \begin{bmatrix} 0.1057 \\ 0.1187 \end{bmatrix} = \begin{bmatrix} 0.33943 \\ 0.33813 \end{bmatrix} )$$

---

This completes the forward and backward propagation steps for a regression problem with one hidden layer and a linear output activation.

# **Example for Binary Clasification Problem**:

###**Setup:**

1. **Input Layer**: 2 neurons
2. **Hidden Layer**: 2 neurons with sigmoid activation
3. **Output Layer**: 1 neuron with sigmoid activation
4. **Loss Function**: Binary Cross-Entropy
5. **Learning Rate**: 0.1

### **Architecture and Components**:

- **Input Layer**: Takes 2 input features.
- **Hidden Layer**: 2 neurons using the sigmoid activation function.
- **Output Layer**: 1 neuron with a sigmoid activation function.

### **Forward Propagation**:

1. **Input Layer to Hidden Layer**:
   - Calculate the weighted sum for each neuron in the hidden layer.
   - Apply the sigmoid activation function:
     $$
     \sigma(x) = \frac{1}{1 + e^{-x}}
     $$
   - The output from each hidden neuron is in the range (0, 1).

2. **Hidden Layer to Output Layer**:
   - Compute the weighted sum of the hidden layer outputs for the output neuron.
   - Apply the sigmoid activation function to this weighted sum to get the final output.

### **Binary Cross-Entropy Loss**:

The Binary Cross-Entropy (BCE) loss function is used for binary classification problems. It measures the performance of a classification model whose output is a probability value between 0 and 1. The formula is:
$$
\text{BCE} = -\frac{1}{N} \sum_{i=1}^{N} [y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i)]
$$
where $(y_i)$ is the true label (0 or 1) and $(\hat{y}_i)$ is the predicted probability for each sample.

### **Backpropagation and Weight Update**:

1. **Calculate Gradients**:
   - Compute the gradient of the Binary Cross-Entropy loss with respect to each weight.
   - For the output layer, the gradient is:
     $$
     \frac{\partial \text{BCE}}{\partial z} = \hat{y} - y
     $$
     where $(z)$ is the weighted input to the output neuron, $(\hat{y})$ is the predicted probability, and $(y)$ is the actual label.

   - For the hidden layer, the gradients are computed by backpropagating the error through the network, taking into account the derivative of the sigmoid function:
     $$
     \sigma'(x) = \sigma(x) \cdot (1 - \sigma(x))
     $$

2. **Update Weights**:
   - Update each weight using the gradient descent rule:
     $$
     w = w - \text{learning rate} \times \frac{\partial \text{BCE}}{\partial w}
     $$

### **Training Process**:

1. **Initialize weights**: Randomly initialize weights.
2. **Forward Propagation**: Compute predictions.
3. **Compute Loss**: Calculate the BCE loss.
4. **Backpropagation**: Compute gradients and update weights.
5. **Repeat**: Iterate through the training data for multiple epochs until the loss converges.

This setup is suitable for binary classification tasks, where the goal is to predict a binary outcome.

## **Forward Propagation**

$$
Z_1 = W_1 \cdot X + b_1

$$
Z_1 = \begin{bmatrix} 0.15 & 0.25 \\ 0.20 & 0.30 \end{bmatrix} \cdot \begin{bmatrix} 0.05 \\ 0.10 \end{bmatrix} + \begin{bmatrix} 0.35 \\ 0.35 \end{bmatrix}

$$
Z_2 = W_2 \cdot A_1 + b_2

$$
Z_2 = \begin{bmatrix} 0.40 \\ 0.45 \end{bmatrix} \cdot \begin{bmatrix} 0.5933 \\ 0.5969 \end{bmatrix} + 0.60

Z_2 = (0.40 \cdot 0.5933) + (0.45 \cdot 0.5969) + 0.60 = 0.7350

Apply the sigmoid activation function:

\hat{y} = \sigma(Z_2) = \frac{1}{1 + e^{-Z_2}} = \frac{1}{1 + e^{-0.7350}} = 0.6752
$$

Use the **Binary Cross-Entropy** loss function:

L = -[y \cdot \log(\hat{y}) + (1 - y) \cdot \log(1 - \hat{y})]

\frac{\partial L}{\partial \hat{y}} = -\frac{1}{0.6752} + \frac{0}{1 - 0.6752} = -1.4802

$$
\frac{\partial L}{\partial Z_1} = \begin{bmatrix} -0.1032 \\ -0.1161 \end{bmatrix} \cdot \begin{bmatrix} 0.2413 \\ 0.2406 \end{bmatrix} = \begin{bmatrix} -0.0249 \\ -0.0279 \end{bmatrix}
